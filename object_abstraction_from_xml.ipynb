```python
# extracting the whole xml in data_images
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()
    
    #extract filename
    
#extract width, height, depths
    image_name = root.find('filename').text if root.find('filename') is not None else None
    width = root.find('size/width').text if root.find('size/width') is not None else None
    height = root.find('size/height').text if root.find('size/height') is not None else None
    
    parser = []
    for obj in root.findall('object'):
        name = obj.find('name').text if obj.find('name') is not None else None
        bndbox = obj.find('bndbox')
        if bndbox is not None:
            xmin = bndbox.find('xmin').text if bndbox.find('xmin') is not None else None
            xmax = bndbox.find('xmax').text if bndbox.find('xmax') is not None else None
            ymin = bndbox.find('ymin').text if bndbox.find('ymin') is not None else None
            ymax = bndbox.find('ymax').text if bndbox.find('ymax') is not None else None
        else:
            xmin = xmax = ymin = ymax = None
        
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])
        
    return parser


    
    

parser_all = list(map(extract_text,xmlfiles))
parser_all
```

```python
data = reduce(lambda x, y : x+y, parser_all)
data
```

```python
df = pd.DataFrame(data, columns=['filename','width','height','name','xmin','xmax','ymin','ymax'])
```

```python
#label encoding 
def label_encoding(x):
    labels = {
        'Cars':0,
        'Bicycle':1,
        'Aeroplane':2,
        'Cats':3,
        'Chair':4,  
        'Dog':5,
        'Bottle':6,
        'Horse':7,
        'Boat':8,
        'Motercycle':9,
        'Cow':10,
        'Sheep':11,
        'Person':12,
        'Potted Plant':13,
        'Sofa':14,
        'Train':15,
        'Television':16,
        'Table':17,
        'Bird':18,
        'Bus':19
    }
 
    
    try:
        return labels[x]
    except:
        #handle case where x is nor found in the labels dict
        return -1 #or any othe appropriate value or action
    
    
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
test_df['id'] = test_df['name'].apply(label_encoding)
```

```python
from shutil import move
```

```python
import os
train_folder = r'Data_images\train'
test_folder = r'Data_images\test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python



def save_data(filename, folder_path, group_obj):
    # Move img
    src = os.path.join('Data_images', filename)
    dat = os.path.join(folder_path, filename)
    print("Source:", src)
    print("Destination:", dat)
    
    try:
        move(src, dat)  # move img to the destined folder
    except FileNotFoundError as e:
        print("Error moving file:", e)
    
    # Save the labels
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0] + '.text')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename, sep=',', index=False, header=False)

# Example usage:
# save_data('001.jpg', 'Data_images\\train', groupby_obj_train)

```

```python
cols = ['filename','id','center_x','center_y','w','h']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data,args=(train_folder,groupby_obj_train))
```

```python
filename_series_test = pd.Series(groupby_obj_test .groups.keys())
```

```python
filename_series_test.apply(save_data,args=(test_folder,groupby_obj_test))
```
